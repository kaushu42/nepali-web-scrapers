```python
from tqdm import tqdm
import utils
```

```python
BASE_URL = 'http://www.majheri.com'
```

```python
files = 1856-9
for page in range(125, 147):
    try:
        print(f'PAGE: {page}')
        url = f'{BASE_URL}/taxonomy/term/6?page={page}'
        soup = utils.get_page_soup(url)
        for heading in tqdm(soup.find_all('h2', {'class': 'title'})):
            if heading.a is None:
                continue
            article_url = f'{BASE_URL}{heading.a["href"]}'
            title = heading.a['title']
            article_soup = utils.get_page_soup(article_url)
            text = article_soup.find_all('div', {'class': 'content'})[4]
            if text is not None:
                text = text.text
            lines = utils.clean(text, keep_numbers=True).split('।')
            with open(f'majheri/{files}.txt', 'w') as f:
                for line in lines:
                    f.write(f'{line}\n')
                files += 1
    except Exception as e:
        print(f'=== ERROR {e} ===')
```
